### Task 1: Load the Excel File

Instructions:

1. Upload the provided excel file into Fabric environment

3. Name the dataframe for sales as df_sales, the dataframe for invoice as df_invoice and the dataframe for product as df_product

```python
### Code Task Here
import pandas as pd

## Create Sales Dataframe
df_sales = pd.read_excel(f"{notebookutils.nbResPath}/builtin/online_sales_retailer 1.xlsx", sheet_name= 'Sales')

display(df_sales.head(5))
```

```python
## Create Invoice Dataframe

df_invoice = pd.read_excel(f"{notebookutils.nbResPath}/builtin/online_sales_retailer 1.xlsx", sheet_name= 'Invoice')

display(df_invoice.head(5))
```

```python
## Create Product dataframe

df_product = pd.read_excel(f"{notebookutils.nbResPath}/builtin/online_sales_retailer 1.xlsx", sheet_name= 'Product')

display(df_product.head(5))
```

### Task 2: Basic Python and Pandas Operations
#### Objective: Perform basic Python and Pandas operations to manipulate the data.

<mark>Instructions:</mark>

1. Check for missing values in the Sales DataFrame.

2. Replace any NaN values in the Sales and UnitPrice columns with 0.

3. Calculate basic descriptive statistics for the Quantity and UnitPrice columns (mean, median, mode, standard deviation).

```python
## Code Task 2 here
# 2.1 Missing values

#    Check for missing values
sales_missing_values = df_sales.isnull()


#   Display only rows with null values
sales_with_nulls = df_sales[sales_missing_values.any(axis=1)]


#   Display the rows with null values
display(sales_with_nulls)


```

```python
#2.2 Replacing null values - note as per above there is no null values so just for reference
# Fill missing values with 0
df_sales_filled = df_sales.fillna({'Sales': 0, 'UnitPrice': 0})
display(df_sales_filled.head(10))
```

```python
#2.3 - statistics for quantity and unit price
mean_quantity = df_sales['Quantity'].mean()
median_quantity = df_sales['Quantity'].median()
mode_quantity = df_sales['Quantity'].mode()
std_quantity = df_sales['Quantity'].std()

print(f"Average Quantity: {mean_quantity}, Median Quantity: {median_quantity}, Mode: {mode_quantity}, Std Dev: {std_quantity}")

mean_unitprice = df_sales['UnitPrice'].mean()
median_unitprice = df_sales['UnitPrice'].median()
mode_unitprice = df_sales['UnitPrice'].mode()
std_unitprice = df_sales['UnitPrice'].std()

print(
    f"Average Unit Price: {mean_unitprice}, Median Unit Price: {median_unitprice}, Mode: {mode_unitprice}, Std Dev: {std_unitprice}")
```

### Task 3: Merging DataFrames
#### Objective: Merge the Sales, Invoice, and Product tables based on common keys to perform a comprehensive analysis.

Instructions:

1. Merge the Sales DataFrame with the Invoice DataFrame on InvoiceNo.

2. Merge the result with the Product DataFrame on StockCode

```python
## Code Task 3 here
#3.1 Merge Sales with Invoice
sales_invoice_merged_df = pd.merge(df_sales, df_invoice, on='InvoiceNo', how = 'left')
display(sales_invoice_merged_df.head(20))

#3.2 Merge result with product
full_merged_df = pd.merge(sales_invoice_merged_df, df_product, on='StockCode', how = 'left')
display(full_merged_df.head(20))
```

## Task 4: Exploratory Data Analysis (EDA)
### Objective: Conduct EDA by analysing the data, identifying patterns, trends, and outliers.

Instructions:

Use describe() to calculate summary statistics for numeric columns.

Perform basic filtering, such as filtering out sales with Quantity equal to 0.

Check for outliers in the Sales column by using a scatter plot.

```python
## Code Task 4 here
print(full_merged_df[['Quantity', 'UnitPrice']].describe())


# Filter out rows where Quantity is 0
df_filtered = full_merged_df[full_merged_df['UnitPrice'] > 0]
display(df_filtered.head(20))

# Scatter plot to visualise the relationship between Quantity and Sales

import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=full_merged_df, x='Quantity', y='UnitPrice', color='green')
plt.title('Sales with outliers')
plt.xlabel('Quantity')
plt.ylabel('Unit Price ($)')
plt.show()
```

## Task 5: Data Visualisation – Scatter Plot, Histogram, and Line Chart
### Objective: Create scatter plots, histograms, and line charts to better understand the data.

Instructions:

Use a scatter plot to examine the relationship between Quantity and Sales.

Use a histogram to visualise the distribution of the Sales column.

Use a line chart to show the trend of Sales over time.


```python
## Code Task 5 here

#New column for sales line total
full_merged_df['Line Total'] = (full_merged_df['Quantity'] * full_merged_df['UnitPrice'])
display(full_merged_df.head())

# group for sales by country
sales_by_country = full_merged_df.groupby('Country')['Line Total'].sum().reset_index()
display(sales_by_country)

# 5.1 - scatterplot

sns.scatterplot(data=full_merged_df, x='Line Total', y='Quantity', color='green')
plt.title('Sales with outliers')
plt.xlabel('Sales')
plt.ylabel('Quantity')
plt.show()

#5.2 histogram
# Histogram to visualise the distribution of Sales - don't find this particularly useful
plt.figure(figsize=(8, 6))
sns.histplot(full_merged_df['Line Total'], bins=10, color='purple', kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Quantity')
plt.show()

#5.3 Line chart
# Convert 'Date' column to datetime format to handle it properly
full_merged_df['InvoiceDate'] = pd.to_datetime(full_merged_df['InvoiceDate'])

# Grouping the data by Date and summing the Sales for each date
sales_by_date = full_merged_df.groupby('InvoiceDate')['Line Total'].sum().reset_index()
display(sales_by_date)

# Creating a line chart to show sales over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='InvoiceDate', y='Line Total', data=sales_by_date, marker='o', color='b')
plt.title('Sales Over Time') 
plt.xlabel('Invoice Date') 
plt.ylabel('Sales') 
plt.xticks(rotation=45)  
plt.show()  




```

## Task 6: Correlation and Data Insights
### Objective: Identify the relationship between different numerical columns in the dataset.

Instructions:

Calculate the correlation between Quantity and Sales.

Visualise the correlation matrix using a heatmap.

```python
## Code Task 6 here
# Calculate correlation between 'Quantity' and 'Sales'
correlation = full_merged_df[['Quantity', 'Line Total']].corr()

# Display the correlation matrix
print(correlation)

# Visualizing the correlation using a heatmap
plt.figure(figsize=(6, 5))  # Set the figure size
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)  # Create the heatmap
plt.title('Correlation Matrix between Quantity and Sales')  # Title of the plot
plt.show()  # Display the plot
```

## Task 7: Grouping and Aggregating
### Objective: Perform grouping and aggregation operations to summarise the data.

Instructions:

GroupBy operation: Group by Product_Category and calculate the total Sales per category.

Create a Pivot Table to show Sales by Country and Product_Category.

```python
## Code Task 7 here
#7.1
# Group by country done above (sales_by_country)
# Display the grouped data
display(sales_by_country)

#7.2
# Create a pivot table to show total sales by Country and Product_Category
pivot_table = full_merged_df.pivot_table(values='Line Total', index='Country', columns='Product_Category', aggfunc='sum')

# Reset the index to make 'StoreCode' a normal column
pivot_table = pivot_table.reset_index()

# Display the pivot table
display(pivot_table)

```
